```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

<h1 style="color:black">Sample Mean</h1>

```python
n=1000
mean=4
var=mean*2
X=np.array([np.random.poisson(mean, 700) for i in range(n)])
sample_mean=np.sum(X,axis=0)/n
plt.figure(figsize=(12,3))
plt.subplot(1,2,1)
plt.title("Approximation of Poisson Dist PDFS",fontsize=16)
plt.ylabel("Density",fontsize=14)
plt.xlabel("x",fontsize=14)
for approx_normal_dist in X[:10]:
    sns.kdeplot(approx_normal_dist)

plt.subplot(1,2,2)
plt.title("sample mean for Normal Dist PDf Of N=10000 ",fontsize=16)
plt.xlabel("x",fontsize=14)
sns.kdeplot(sample_mean,shade=True)
plt.show()

table = pd.DataFrame({'X':[4, 8], 'Sample Mean':[round(np.mean(sample_mean), 2), round(np.var(sample_mean), 2)]})
table.rename(index={0:'mean', 1:'variance'}, inplace=True)
print(table)
```

<h1> Gaussian Dist

```python
mean=5
var=25
X=np.random.normal(mean,math.sqrt(var),2000)
Y=1.5*X+4
plt.figure(figsize=(8,3))
plt.title("approximation Normal PDF before and after transforamtion",fontsize=16)
plt.ylabel("density",fontsize=14)
plt.xlabel("x",fontsize=14)
sns.kdeplot(X,shade=True,label="X")
sns.kdeplot(Y,shade=True,label="Y=3*X+4",lw=3)
plt.xlim(-15,30)
plt.legend(fontsize=14)
plt.show()

table = pd.DataFrame({'X':[mean, var], 'Y':[round(np.mean(Y), 2), round(np.var(Y), 2)]})
table.rename(index={0:'mean', 1:'variance'}, inplace=True)
print(table)
```

<h1>Standrize Gaussian 

```python
Z=(X-mean)/math.sqrt(var)

plt.figure(figsize=(8,3))
plt.title("approximation Normal PDF before and after Standrization",fontsize=16)
plt.ylabel("density",fontsize=14)
plt.xlabel("x",fontsize=14)
sns.kdeplot(X,shade=True,label="X")
sns.kdeplot(Z,shade=True,label="Z",lw=3)
plt.xlim(-15,30)
plt.legend(fontsize=14)
plt.show()

table = pd.DataFrame({'X':[mean, var], 'Z':[round(np.mean(Z), 2), round(np.var(Z), 2)]})
table.rename(index={0:'mean', 1:'variance'}, inplace=True)
print(table)
```

<h1> Law of Large Number (LLN)</h1>
<li> The average of N random variables tend to close the expected value if N goes to infinty</li>
<h1>Strong Law Of Large Numbers</h1>
<li>The sample mean converge almost surley to the expected value  as N goes to infinty</li>

```python
X=np.random.normal(0,1,100000)
sample_sizes=[50000,60000,70000,80000,90000,1000000]
plt.title("Sample Size to infinity , Variance to 0",fontsize=16)
plt.ylabel("density",fontsize=14)
for n_sample in sample_sizes:
    sample=X[:n_sample]
    sns.kdeplot(sample/n_sample,label=f'n={n_sample:,}')
    
```

<h1>Weak Law Of Large Number</h1>
<img src="Imgs/Weak_LLN.png">

```python
X=np.array([np.random.normal(2,5,1000) for i in range(200)])
sample_sizes=[1,20,150,200]
plt.figure(figsize=(20,4))
for i in range(4):
    plt.subplot(1, 4, i + 1)
    plt.title(f'n={sample_sizes[i]}')
    sns.kdeplot(np.sum(X[:sample_sizes[i]],axis=0)/sample_sizes[i])
            
```
